```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the original image
image_path = 'C:/Users/navee/OneDrive/Desktop/Minor-2/Eye_diseases_dataset/Eye_diseases_dataset/3_retina_disease/Retina_066.png'
original_image = cv2.imread(image_path)
original_image = cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB

# Convert the original image to grayscale
gray_image = cv2.cvtColor(original_image, cv2.COLOR_RGB2GRAY)

# Find edges in the grayscale image
edges = cv2.Canny(gray_image, 50, 150)

# Threshold the edge image to get a binary mask
_, binary_mask = cv2.threshold(edges, 0, 255, cv2.THRESH_BINARY)

# Apply the binary mask to the original image to highlight the edges
edge_highlight = cv2.bitwise_and(original_image, original_image, mask=binary_mask)

# Save the black and white image
cv2.imwrite('black_and_white_image.png', gray_image)

# Save the edge-highlighted image
cv2.imwrite('edge_highlighted_image.png', cv2.cvtColor(edge_highlight, cv2.COLOR_RGB2BGR))

# Display the images
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# Display black and white image
axs[0].imshow(gray_image, cmap='gray')
axs[0].set_title('Black and White Image')
axs[0].axis('off')

# Display edge-highlighted image
axs[1].imshow(edge_highlight)
axs[1].set_title('Edge Highlighted Image')
axs[1].axis('off')

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the original image
image_path = 'C:/Users/navee/OneDrive/Desktop/Minor-2/Eye_diseases_dataset/Eye_diseases_dataset/3_retina_disease/Retina_066.png'
original_image = cv2.imread(image_path)
original_image_rgb = cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB

# Convert the original image to grayscale
gray_image = cv2.cvtColor(original_image, cv2.COLOR_BGR2GRAY)

# Apply histogram equalization to improve contrast
equalized_image = cv2.equalizeHist(gray_image)

# Apply adaptive thresholding
binary_image = cv2.adaptiveThreshold(equalized_image, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 2)

# Find unchanged areas by comparing the black and white image with the grayscale image
unchanged_areas = cv2.bitwise_and(equalized_image, binary_image)

# Create a binary mask for highlighting unchanged areas
highlight_mask = np.zeros_like(original_image_rgb)
highlight_mask[unchanged_areas != 0] = [255, 0, 0]  # Highlight unchanged areas in red

# Add the highlight mask to the original RGB image
highlighted_image = cv2.addWeighted(original_image_rgb, 0.7, highlight_mask.astype(np.uint8), 0.3, 0)

# Display the original input image, black and white image, and the original image with unchanged areas highlighted
fig, axs = plt.subplots(1, 3, figsize=(18, 6))

# Display original input image
axs[0].imshow(original_image_rgb)
axs[0].set_title('Original Input Image')

axs[0].axis('off')

# Display black and white image
axs[1].imshow(binary_image, cmap='gray')
axs[1].set_title('Improved Black and White Image')
axs[1].axis('off')

# Display original image with unchanged areas highlighted
axs[2].imshow(highlighted_image)
axs[2].set_title('Original Image with Unchanged Areas Highlighted')
axs[2].axis('off')

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load OCT image
image_path = 'C:/Users/navee/OneDrive/Desktop/Minor-2/OCT2017/test/NORMAL/NORMAL-76914-1.jpeg'
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Check if image is loaded successfully
if image is None:
    print("Error: Unable to load image")
else:
    # Segmentation
    # Example: Simple thresholding to segment layers
    _, binary_image = cv2.threshold(image, 100, 255, cv2.THRESH_BINARY)

    # Find contours
    contours, _ = cv2.findContours(binary_image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Get the contour with the maximum area (assuming it's the peak)
    max_contour = max(contours, key=cv2.contourArea)

    # Calculate area of the peak in pixels
    peak_area_pixels = cv2.contourArea(max_contour)

    # Define conversion factor from pixels to desired units (e.g., cm^2)
    pixel_to_cm2 = (0.1) ** 2 

    # Convert area to square centimeters
    peak_area_cm2 = peak_area_pixels * pixel_to_cm2

    # Draw the contour (peak) on the original image
    result_image = cv2.cvtColor(image, cv2.COLOR_GRAY2BGR)
    cv2.drawContours(result_image, [max_contour], -1, (0, 255, 0), 2)

    # Display both original and segmented images with the peak highlighted
    plt.figure(figsize=(15, 5))

    plt.subplot(1, 2, 1)
    plt.imshow(image, cmap='gray')
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(result_image)
    plt.axis('off')

    # Add text below the images
    plt.figtext(0.5, 0.01, f"Area of the highlighted region: {peak_area_cm2:.2f} cm^2", ha="center", fontsize=12)

    # Adjust layout to make room for the text
    plt.subplots_adjust(bottom=0.05)

    plt.show()
```
